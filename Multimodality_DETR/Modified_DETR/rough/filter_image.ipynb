```python
#***********************************Test_visible*********************************************

import json
import os
import shutil

# Load the JSON file
with open('test_annotations.json') as f: #************************change
    data = json.load(f)

# Get the list of images from the JSON file
image_list = [img['file_name'] for img in data['images']]

# Define the source and destination directories
# src_dir = r"C:\Users\aland\Downloads\kaist-cvpr15\images" # Update this path
src_dir = "/data/Singlemodality_DETR_V1/kaist-cvpr15/images/"
# dst_dir = r"C:\Users\aland\Downloads\selected_images_visible" # Update this path
dst_dir = "/data/DETR/dataset_visible/test"

# Create the destination directory if it doesn't exist
os.makedirs(dst_dir, exist_ok=True)

# Iterate over the image list
for image_name in image_list:
    # Extract the set and V number from the image name
    set_num, v_num, _ = image_name.split('_')

    # Define the source path
    src_path = os.path.join(src_dir, set_num, v_num, 'visible', image_name)

    # Check if the source file exists
    if not os.path.exists(src_path):
        # print("path error ")
        continue

    # Define the destination path
    dst_path = os.path.join(dst_dir, image_name)

    # Try to move the image
    try:
        shutil.copy(src_path, dst_path)
        print(f'copied file {src_path} to {dst_path}')
    except Exception as e:
        print(f'Failed to copy file {src_path} to {dst_path}: {e}')
```
